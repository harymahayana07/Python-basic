```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Data/cereal.csv")
df
```

```python
df_bar = df.groupby('mfr').count()['name'].reset_index()
```

```python
df_bar
```

```python
# barplot
plt.figure(figsize=(5,5))
sns.barplot(y="mfr",x="name",data=df_bar)
plt.title("MFR")
```

```python
# count plot
sns.countplot(x='mfr',data=df)
```

```python
df_stacked = df.groupby('mfr').mean().reset_index()
df_stacked
```

```python
# stacked bar chart
plt.figure(figsize=(5, 5))
bar_protein = sns.barplot(x="mfr", y="protein", data=df_stacked,color="darkblue")
bar_fat = sns.barplot(x="mfr", y="fat", data=df_stacked,color="lightblue")
bar_potass = sns.barplot(x="mfr", y="potass", data=df_stacked,color="")

plt.title("MFR")
```

```python
sns.distplot(df['rating'])
```

```python
# heatmap
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),annot=True,cmap="Blues")
```

```python
# numerik = angka angka
sns.scatterplot(x="sugars",y="rating",data=df)
```

```python
sns.boxplot(x="mfr",y="rating",data=df)
```

```python
# catplot
sns.catplot(x="mfr",y='rating',data=df)
```

```python
# implot
sns.lmplot(x="sugars",y="rating",data=df)
```

```python
sns.jointplot(x="calories",y="rating",data=df,kind='reg')
```
