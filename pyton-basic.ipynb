```python
# Belajar Python Basic Part1

# library import
import time
from numpy import var

# inisialisasi variabel & implement operator aritmatika
var = "buah"
a = 4 + 2
b = 4 - 2
c = 4 * 2
d = 4 / 2
e = 5 % 2
f = 5**2
g = 5//2
h = -5//2
print(var)
print(a)
print(b)
print(c)
print(d)
print(e)
print(f)
print(g)
print(h)

# relational Operator
i = 5 < 2
j = 5 > 2
k = 5 <= 2
l = 5 >= 2
m = 5 == 2
n = 5 != 2
# logical Operator
op = 5
qr = 2
st = 3
if qr<op and op>st:
    print("And")

if op<qr or op>st:
    print("Or")
if not(5<2):
    print("Not")

print(i)
print(j)
print(k)
print(l)
print(m)
print(n)

# list
daftar_belanja = ["ayam","daging","tempe"]
print("Arik Membeli "+daftar_belanja[0]+" dan "+daftar_belanja[1]+" dipasar")

# manipulasi list
daftar_belanja[0] = "kangkung"
print(daftar_belanja)

# function in list
print(len(daftar_belanja)) 
daftar_pembelian= daftar_belanja.copy()
print(daftar_pembelian)
# append untuk menambah list di akhir
daftar_pembelian.append("telur")
print(daftar_pembelian)
# remove() untuk hapus
daftar_pembelian = ["kangkung", "daging", "tempe", "telur"]
daftar_pembelian.remove("telur")
print(daftar_pembelian)
# tuple bersifat permanen, berbeda dgn list(tidak permanen)
daftar_belanja = ("ayam","daging","tempe")
print(daftar_belanja)
# Set atau himpunan
himpunan = {4, 5, 6, 2}
print(himpunan)
# remove set
himpunan.remove(5)
print(himpunan)

# Dictionary
numbers = {'one':1,'two':2,'three':3}
print(numbers['one'])

# definisi function in python
def matematika(ar,ik,kk):
    penjumlahan = ar + ik + kk
    pengurangan = ik - kk - ar
    pembagian = ar/ik
    return penjumlahan,pengurangan,pembagian

# default parameter
def greet():
    print("Hallo. Hary")
def greet(who = "Arik"):
    print("Hello,",who)

# keyword parameter
def kenalan(nama):
    print(nama)
kenalan(nama = "Hary")

# looping
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
for planet in planets:
    print(planet)
# looping for range
for i in range(5):
    print("Hary mahayana")
time.sleep(1)
# break / while
while True:
    a = input('input string: ')
    if a == 'quit':
        break
    print('length of a is {}'.format(len(a)))
#implement continue 
fruits = ['apple', 'banana', 'orange', 'grape', 'pineapple']
for fruit in fruits:
  if fruit == 'orange':
    continue
  print('my fruit is {}'.format(fruit))
  



```
